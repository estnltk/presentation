## Laused isikunimedega test5

Andmebaasi tabelist loetakse sisse lausete veerg.

```python
from estnltk import Text
import sqlite3
import pickle
```

```python
corpus = []
```

```python
con = sqlite3.connect("data.db")
cur = con.cursor()
```

```python
for row in cur.execute("SELECT sentences FROM data"):
    row = pickle.loads(row[0])
    corpus.append(row)
```

```python
con.close()
```

Selleks, et isikunime märgendit sisaldavad laused andmebaasi lisada, kogutakse need kahemõõtmelisse järjendisse, sest ühe repliigi kohta võib isikunime sisaldavaid lauseid olla rohkem kui üks.

```python
nested_sentences_with_per = []
```

```python
for sentences in corpus:
    for sentence in sentences:
        if 'PER' not in sentence.ner.nertag:
            sentences.remove(sentence)
    nested_sentences_with_per.append(sentences)
```

```python
len(nested_sentences_with_per)
```

```python
display(nested_sentences_with_per)
```

Tekitan igaks juhuks ka ühemõõtmelise järjendi kõigist lausetest:

```python
sentences = [sentence for sentences in corpus for sentence in sentences]
```

```python
print(len(corpus))
print(len(sentences))
```

Seejärel tekitan igaks juhuks ühemõõtmelise järjendi isikunime märgendit sisaldavatest lausetest:

```python
sentences_with_per = [sentence for sentence in sentences if 'PER' in sentence.ner.nertag]
```

```python
len(sentences_with_per)
```


Laused kahemõõtmelises järjendis, kus leidub isikunime märgend lisatakse andmebaasi tabelis uude veergu neile vastavate repliikide kohale.

```python
con = sqlite3.connect("data.db")
cur = con.cursor()
cur.execute('''ALTER TABLE data ADD COLUMN sentences_with_per BLOB''')
```

```python
for sentences in nested_sentences_with_per:
    if len(sentences) > 0:
        sen_bytes = pickle.dumps(sentences)
        cur.execute("""UPDATE data SET sentences_with_per = ? WHERE number = ? AND person = ? AND date = ?""", (bytearray(sen_bytes), sentences[0].meta['line_number'], sentences[0].meta['person'], sentences[0].meta['date']))
        con.commit()
```
