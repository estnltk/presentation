## PhraseTagger nimeüksuste tuvastamiseks test5

```python
from estnltk import Text
from estnltk.taggers.system.rule_taggers import StaticExtractionRule, Ruleset
from estnltk.taggers import PhraseTagger
import csv
import sqlite3
from estnltk.converters import json_to_text
```

Notebookis 04 sai loodud loend isiku- ja organisatsiooninimeüksustest. Seda loendit saab kasutada ära järgnevalt PhraseTaggeri töös, et leida üles tekstist need nimeüksused, mis NerTaggeril tuvastamata jäid.

```python
vocabulary_file = 'entity_phrases.csv'
ruleset = Ruleset()
ruleset.load(file_name=vocabulary_file, key_column='phrase')
```

```python
#dekoraator
```

```python
phrasetagger = PhraseTagger(output_layer='entity_phrases',
                      input_layer='morph_analysis',
                      input_attribute='lemma',
                      ruleset=ruleset,
                      output_attributes=['value'],
                      #decorator=decorator,
                      conflict_resolver='KEEP_MAXIMAL',
                      ignore_case=True)
```

Esialgu katsetan taggeri tööd mõne repliigi peal, sest tulemus peaks olema ilmekam. Hiljem märgendan sellega lauseid.

```python
test_lines = []
```

```python
con = sqlite3.connect("media_data.db")
cur = con.cursor()
```

```python
for row in cur.execute("SELECT line FROM lines WHERE ID >= 6 AND ID <= 12"):
    line = json_to_text(json_text=row[0])
    test_lines.append(line)
```

```python
con.close()
```

```python
for line in test_lines:
    display(line)
    display(line.ner)
```

```python
for line in test_lines:
    phrasetagger.tag(line)
```

```python
for line in test_lines:
    display(line.ner, line.entity_phrases)
```
