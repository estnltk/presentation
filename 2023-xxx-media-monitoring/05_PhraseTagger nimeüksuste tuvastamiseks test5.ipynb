## PhraseTagger nimeüksuste tuvastamiseks test5

```python
from estnltk import Text
from estnltk.taggers.system.rule_taggers import AmbiguousRuleset, Ruleset, StaticExtractionRule
from estnltk.taggers import PhraseTagger
import csv
import sqlite3
from estnltk.converters import json_to_text
from estnltk.converters import text_to_json
```

Notebookis 04 sai loodud loend isiku- ja organisatsiooninimeüksustest. Seda loendit saab kasutada ära järgnevalt PhraseTaggeri töös, et leida üles tekstist need nimeüksused, mis NerTaggeril tuvastamata jäid.

```python
vocabulary_file = 'entity_phrases.csv'
ruleset = AmbiguousRuleset()
ruleset.load(file_name=vocabulary_file, key_column='phrase')
```

Panen paika kaks esmast reeglit nimede käänamise kohta:

1. Kui tegemist on isikunimeüksusega, mille pikkus on > 1, siis juhul kui selle nimeüksuse ükskõik millisel positsioonil olev osa (välja arvatud osa positsioonil -1) on millegis muus, kui nimetavas käändes, ei kuulu sellele osale järgnevad osad tegelikult selle nimeüksuse koosseisu.

2. Kui tegemist on organisatsiooninimeüksusega, mille pikkus on > 1, siis juhul kui selle nimeüksuse osa positsioonil -2 on millegis muus, kui omastavas käändes, ei kuulu osa positsioonil -1 kindlasti nimeüksuse koosseisu.

```python
#dekoraator
```

```python
phrasetagger = PhraseTagger(output_layer='entity_phrases',
                      input_layer='morph_analysis',
                      input_attribute='lemma',
                      ruleset=ruleset,
                      output_attributes=['value'],
                      #decorator=decorator,
                      conflict_resolver='KEEP_MAXIMAL',
                      ignore_case=True)
```

Andmebaasist loetakse sisse kõik laused, et märgendada PhraseTaggeri abil uuesti peale nimistus olevad nimeüksused.

```python
sentences = []
sentence_ids = []
```

```python
con = sqlite3.connect("media_data.db")
cur = con.cursor()
```

```python
for row in cur.execute("SELECT ID, sentence FROM sentences"):
    sentence_id = row[0]
    sentence = json_to_text(json_text=row[1])
    sentences.append(sentence)
    sentence_ids.append(sentence_id)
```

```python
con.close()
```

```python
for sentence in sentences:
    phrasetagger.tag(sentence)
```

Ilmselt ülekattumisvea tõttu tekib veateade, mistõttu püüan erindist mööda minna:

```python
for sentence in sentences:
    try:
        phrasetagger.tag(sentence)
    except:
        pass
```

Töötab...

Vaatame üht lauset, kus tagger mõne nimeüksuse tuvastas:

```python
for sentence in sentences:
    if len(sentence.entity_phrases) > 0:
        display(sentence, sentence.entity_phrases)
        break
```

Kuna lausete seas on lauseid, millele ei sobi taggeri konfliktilahendusstrateegia (maksimaalse vaste säilitamine) püüan selgitada järgnevalt välja, millistes lausetes probleem esineb:

```python
for sentence in sentences:
    try:
        len(sentence.entity_phrases) > 0
    except:
        display(sentence)
```

Andmebaasi luuaks uus tabel lausete jaoks, milles PhraseTagger mõne nimeüksusefraasi tuvastas.

```python
con = sqlite3.connect("media_data.db")
cur = con.cursor()
```

```python
cur.execute("CREATE TABLE phrase_tagger_sentences(ID INTEGER PRIMARY KEY, sentence_ID INTEGER, sentence TEXT)")
```

```python
for i in range(len(sentences)):
    try:
        if len(sentences[i].entity_phrases) > 0:
            sentence_json = text_to_json(sentences[i])
            cur.execute("""INSERT INTO phrase_tagger_sentences
                                    (sentence_ID, sentence)
                                    VALUES (?, ?);""", (sentence_ids[i], sentence_json))
    except:
        continue
    con.commit()
```

```python
con.close()
```
