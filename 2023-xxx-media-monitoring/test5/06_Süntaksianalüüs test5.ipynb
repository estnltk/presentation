## Süntaksianalüüs test5

```python
from estnltk import Text
from estnltk.taggers import StanzaSyntaxEnsembleWebTagger
from estnltk_neural.taggers.syntax.stanza_tagger.ensemble_tagger import StanzaSyntaxEnsembleTagger
from estnltk.taggers import SyntaxDependencyRetagger
import sqlite3
from estnltk.converters import json_to_text
import csv
from collections import Counter
```

Veebi kaudu kasutatav tagger:

```python
stanza_syntax_ensemble_web_tagger = \
    StanzaSyntaxEnsembleWebTagger(url='https://api.tartunlp.ai/estnltk/tagger/stanza_syntax')
stanza_syntax_ensemble_web_tagger
```

Päris:

```python
stanza_ensemble_tagger = StanzaSyntaxEnsembleTagger(output_layer='stanza_ensemble_syntax',
                                            input_morph_layer='morph_extended', add_parent_and_children=True)
```

Kõigepealt loetakse andmebaasist sisse laused, millele on PhraseTaggeriga märgendatud peale mõni huvipakkuvatest nimeüksustest.

```python
sentences = []
```

```python
con = sqlite3.connect("media_data.db")
cur = con.cursor()
```

```python
for row in cur.execute("SELECT sentence FROM phrase_tagger_sentences"):
    sentence = json_to_text(json_text=row[0])
    sentences.append(sentence)
```

```python
con.close()
```

```python
len(sentences)
```

Seejärel märgendatakse lausetele süntaksianalüüsi kiht.

```python
#Veebitaggeriga
for sentence in sentences:
    sentence.tag_layer('morph_extended')
    stanza_syntax_ensemble_web_tagger.tag(sentence)
    SyntaxDependencyRetagger('stanza_ensemble_syntax').retag(sentence) #Parent ja children atribuutide lisamiseks kihile
```

Veebitaggeriga tekkis märgendamisel mõne lause juures viga, mis on seotud PhraseTaggeri abil peale märgendatud nimeüksuste kihiga.

```python
#Päris taggeriga
for sentence in sentences:
    sentence.tag_layer('morph_extended')
    stanza_ensemble_tagger.tag(sentence)
```

Nimeüksuse span/spanid leitakse entity_phrases kihist:

```python
phrase = []
```

```python
for entity in sentences[14].entity_phrases:
    for span in entity:
        phrase.append(span)
```

```python
print(phrase)
```

Leitakse nimefraasi ülemus (juur), mis peab olema sõnaliigilt verb.

```python
for subspan in phrase:
    parent = sentences[14].stanza_ensemble_syntax.get(subspan)['parent_span']
    parent_word = text['words'].get(parent) 
    if parent_word in phrase:
        continue
    
    # Juur
    parent_morph = sentences[14].morph_analysis.get(parent)
    if 'V' not in parent_morph['partofspeech']:
        continue
    
    print(list(s.text for s in phrase))
    print(parent['lemma'])
```

Seejärel leitakse nimefraasi alamad (siin otsitakse ainult nimisõna(fraasi) täiendeid tähistusega "amod").

```python
for subspan in phrase:
    children = sentences[14].stanza_ensemble_syntax.get(subspan)['children']
    if len(children) > 0:
        for child in children:
            child_word = sentences[14]['words'].get(child) 
            if child_word in phrase:
                continue
            elif 'amod' != child['deprel']:
                continue
            print(child)
```

Eelnevas lauses ei olnud nimeüksusel täiendeid, mistõttu alamaid ei tagastatud. Järgnevalt võetakse ette kõik laused. Kõigepealt luuakse kaks tühja järjendit vastavalt kas täiendite või verbide (juurte) jaoks.

```python
amods = []
verbs = []
```

```python
for sentence in sentences:
    for entity in sentence.entity_phrases:
        phrase = []
        for span in entity:
            phrase.append(span)
    
        for subspan in phrase:
            parent = sentence.stanza_ensemble_syntax.get(subspan)['parent_span']
            if parent != None:
                parent_word = sentence['words'].get(parent) 
                if parent_word in phrase:
                    continue
                # Juur
                parent_morph = sentence.morph_analysis.get(parent)
                if 'V' not in parent_morph['partofspeech']:
                    continue
                verbs.append(parent['lemma'])
                print(list(s.text for s in phrase), parent['lemma'])
            
            # Täiendid
            children = sentence.stanza_ensemble_syntax.get(subspan)['children']
            if len(children) > 0:
                for child in children:
                    child_word = sentence['words'].get(child) 
                    if child_word in phrase:
                        continue
                    elif 'amod' != child['deprel']:
                        continue
                    amods.append(child['lemma'])
                    print(child['lemma'], list(s.text for s in phrase))
```

```python
# Täiendid
print(amods)
```

```python
# Verbid
print(verbs)
```

Järgnevalt tehakse leitud sõnadest sagedusloendid, mis salvestatakse CSV-faili.

```python
frequencies_amods = Counter(amods)
frequencies_verbs = Counter(verbs)
```

```python
with open('verbs_amods_frequencies.csv', 'w', newline='', encoding='UTF-8') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=['lemma', 'freq', 'value'])
    writer.writeheader()
    header2 = {'lemma': 'string', 'freq': 'integer', 'value': 'string'}
    writer.writerow(header2)
    for lemma in frequencies_amods.keys():
        writer.writerow({'lemma': lemma, 'freq': frequencies_amods[lemma], 'value': 'amod'})
    for lemma in frequencies_verbs.keys():
        writer.writerow({'lemma': lemma, 'freq': frequencies_verbs[lemma], 'value': 'verb'})
```

```python
with open('verbs_amods_frequencies.csv', encoding='UTF-8') as csv_file:
    rows = []
    reader = csv.DictReader(csv_file)
    for row in reader:
        rows.append(row)
```
