```python
from estnltk import Text
from estnltk.taggers.system.rule_taggers import AmbiguousRuleset, Ruleset, StaticExtractionRule
from estnltk.taggers import PhraseTagger
import csv
```

```python
#prooviks kaks reeglit
phrase_list = [StaticExtractionRule(pattern=('aadu','must')),
               StaticExtractionRule(pattern=('euroopa','liit')),
              ]
test_ruleset = Ruleset()
test_ruleset.add_rules(phrase_list)
```

```python
#prooviks kaks teksti. Esimeses kuuluvad mõlemad nimeüksused täies pikkuses märgendamisele, teises mitte
text1 = Text('Täna räägime Aadu Mustast ja Euroopa Liidust.')
text2 = Text('Eile rääkisime Aadule musta kaabu kinkimisest ja Euroopas liidu sõlmimisest.')

#mõlemale tekstile märgitakse peale morfoloogilise analüüsi kiht
text1.tag_layer('morph_analysis')
text2.tag_layer('morph_analysis')

#vaatame, esimese teksti morfoloogilise analüüsi kihti
text1.morph_analysis
```

```python
#ja teise teksti morfoloogilise analüüsi kihti
text2.morph_analysis
```

```python
#dekoraatori katsetus
def decorator1(text, span, annotation):
    print(text.morph_analysis[span[0]]) #Leitud fraasist esimese sõna Span-objekt
    print(text.morph_analysis[span[0]].form) #Leitud fraasist esimese sõna sõnavorm
    print(span[0]) #Leitud fraasi esimene Span
    print(annotation['phrase'][0]) #Leitud fraasi esimene sõna
```

```python
TestTagger1 = PhraseTagger(output_layer='entities',
                         input_layer='morph_analysis',
                         input_attribute='lemma', 
                         ruleset=test_ruleset,
                         output_attributes=('match', 'value'),
                         decorator=decorator1,
                         conflict_resolver='KEEP_MAXIMAL',
                         ignore_case=True)
```

```python
TestTagger1.tag(text1)
```

```python
TestTagger1.tag(text2)
```

Dekoraatoriga on võimalik leida tuvastatud fraasi kõigi sõnade käänded, mis on aluseks valete käändekombinatsioonide tuvastamisel ja eemaldamisel.

```python
#Teine dekoraatori katsetus
def decorator1(text, span, annotation):
    if len(annotation['phrase']) > 1:
        i = len(annotation['phrase'])-2
        index_for_removal = 0
        while i > 0:
            if text.morph_analysis[span[i]].form != 'sg n' or text.morph_analysis[span[i]].form != 'pl n':
                index_for_removal = i
                i-=1
            else:
                i-=1
        if index_for_removal != 1:
            return # Siin jäi pooleli
        
    else:
        return annotation
```
